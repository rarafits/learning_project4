# Soal 2

Paired T-test adalah varian dari T-test dimana sebuah sampel akan diukur sebanyak 2 kali (biasanya sebelum vs setelah suatu *treatment*).

Lakukan uji hipotesis apakah nilai ujian murid menjadi lebih baik setelah pemberian modul atau tidak.

Gunakan code dari dokumentasi berikut
https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html

```python
# A> Tulis H0 dan H1
# H0 : Nilai ujian murid tidak menjadi lebih tinggi setelah pemberian modul
# H1 : Nilai ujian murid menjadi lebih tinggi setelah pemberian modul

# B> Tulis kode untuk melakukan test
from scipy.stats import ttest_rel

# Menghitung nilai T dan p-value
T, p_value = ttest_rel(df['Pre-module'], df['Post-module'])

# Menampilkan hasil uji hipotesis
alpha = 0.05
print(f"T-score: {T:.2f}")
print(f"p-value: {p_value:.5f}")
if p_value < alpha:
    print("Ada bukti menolak H0. Nilai ujian murid menjadi lebih tinggi setelah pemberian modul.")
else:
    print("Tidak ada bukti untuk menolak H0. Nilai ujian murid tidak menjadi lebih tinggi setelah pemberian modul.")
```

#C> TULIS KESIMPULAN ANDA
Nilai p-value (0.00439) < alpha (0.05), maka kita dapat menolak hipotesis nol (H0) dan menyimpulkan bahwa terdapat bukti bahwa nilai ujian murid menjadi lebih baik setelah pemberian modul.
